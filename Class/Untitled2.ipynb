```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(y,z)
plt.show()
```

```python
plt.plot(z,y)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,z,'g-*')
plt.subplot(2,2,3)
plt.plot(x,y,'b+-')
plt.subplot(2,2,4)
plt.plot(y,z,'y+-')
```

```python
plt.plot(x,y)
plt.xlabel('Angles')
plt.ylabel('Sine Wave')
plt.title('Sine Wave Curve')
plt.show()
```

```python
x = np.linspace(20,40,15)
y = 3*x + 5
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Linear curve')
plt.show()
```

```python
x = np.linspace(-10,10,1000)
y = x**2
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Parabola Curve')
plt.show()
```
